## 문제 

<p>정수를 저장하는 큐를&nbsp;구현한 다음, 입력으로 주어지는 명령을 처리하는 프로그램을 작성하시오.</p>

<p>명령은 총 여섯 가지이다.</p>

<ul>
<li>push X: 정수 X를 큐에 넣는 연산이다.</li>
<li>pop: 큐에서 가장 앞에 있는 정수를 빼고, 그 수를 출력한다. 만약 큐에 들어있는 정수가 없는 경우에는 -1을 출력한다.</li>
<li>size: 큐에 들어있는 정수의 개수를 출력한다.</li>
<li>empty: 큐가&nbsp;비어있으면 1, 아니면 0을 출력한다.</li>
<li>front: 큐의 가장 앞에 있는 정수를 출력한다. 만약 큐에 들어있는 정수가 없는 경우에는 -1을 출력한다.</li>
<li>back:&nbsp;큐의 가장 뒤에 있는 정수를 출력한다. 만약 큐에 들어있는 정수가 없는 경우에는 -1을 출력한다.</li>
</ul>

```python
import sys

class My_queue():
    def __init__(self):
        self.queue = []
        
    def push(self, num):
        self.queue.append(num)
        
    def pop(self):
        if len(self.queue):
            return print(self.queue.pop(0))

        else : return print(-1)

    def empty(self):
        if len(self.queue): return print(0)

        else: return print(1)

    def front(self):
        if len(self.queue):
            return print(self.queue[0])

        else : return print(-1)

    def back(self):
        if len(self.queue):
            return print(self.queue[-1])

        else : return print(-1)

    def size(self):
        return print(len(self.queue))

```

```python
cmd_num = int(sys.stdin.readline().rstrip())

queue = My_queue()

for i in range(cmd_num):
    cmd = sys.stdin.readline().rsplit()

    if cmd[0] == "push":
        queue.push(cmd[1])

    elif cmd[0] == "empty":
        queue.empty()

    elif cmd[0] == "pop":
        queue.pop()
    
    elif cmd[0] == "front":
        queue.front()

    elif cmd[0] == "back":
        queue.back()

    elif cmd[0] == "size":
        queue.size()
```
